# Calculating Across Rows

How do we calculate row-wise in Pandas like we do in Excel?

<img src="https://sandeepmj.github.io/image-host/oil-v-gold.png">


We need a couple of new concepts to do this calculation:

1. Capture a number from a cell and hold in variable.
2. We `apply()`calculations across columns, row-wise, using using a `lambda expression`.


This is step one in your ability to accomplish tasks that are highly specialized to your beats.



````python
## import library
import pandas as pd
````

````python
## pull in data
df = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/refs/heads/main/oil-v-gold-pandas.csv")
df
````

What are the different values for row **TWO** in this formula for `PX_GOLD`:

$$\text{Pct Chg at time } t = \frac{\text{Value at time } t - \text{Value at start time}}{\text{Value at start time}} \times 100$$

Use `.iat` to store a value from a specific cell into a variable:

```dataframe["column_name"].iat[row_index_number]```

````python
## hold gold

df["PX_GOLD"].iat[0]
````

Also works:

`dataframe.iat[col_index_number, row_index_number]`

````python
## hold gold col index, row index
gold_baseline = df.iat[0,1]
gold_baseline
````

Place the value for the starting oil price in a variable called `oil_baseline`

````python
## oil_baseline
oil_baseline = df.iat[0,2]
oil_baseline
````

# `lambda expressions`

We're going to just write one first and then break it down:

````python
## create a new column that holds the zero index for gold
df["gold_zero_index"] = df["PX_GOLD"].apply(lambda x: (x-gold_baseline)/gold_baseline * 100)
````

````python
## call df
df
````

## What  `x` stand for:

- In Pandas, we `apply()` lambda expressions to columns.
-  `.apply()`  works element-wise on each value in the `PX_GOLD` column, one at a time.
- `x` is a single value representing a cell in the column `PX_GOLD`
- The ```lambda x:``` processes each row of a dataframe, one row at a time.
-  The `lambda expression` takes each individual number `x` from `PX_GOLD` and runs the indexing formula on it.

Think of it as: "call each value in `PX_GOLD` as `x`, and do this math to using `x` in the formula." 

## create a new column that holds the zero index for oil



````python
## create a new column that holds the zero index for oil

df["oil_zero_index"] = df["PX_OIL"].apply(lambda x: (x-oil_baseline)/oil_baseline * 100)
df
````

### 100-indexing

Create cells below to add columns for both oil and gold for 100-indexing.

Reminder, the formula for 100-indexing:

$$\text{Index} = \left(\frac{\text{value at time } t}{\text{value at start}}\right) \times 100$$

````python
## gold 100
df["gold_100_index"] = df["PX_GOLD"].apply(lambda x: round(x/gold_baseline * 100, 1))
df
````

````python
## oil 100
df["oil_100_index"] = df["PX_OIL"].apply(lambda x: (x)/oil_baseline * 100)
df
````

<a href="https://raw.githubusercontent.com/sandeepmj/datasets/refs/heads/main/demo-data-solutions/cablebill-v-income.csv">This dataset</a> shows the price of a standard cable subscription and the U.S. median income annually between 1995-2023.

Has median income kept pace with the price of a cable subscription?

Round all numbers to 1 decimal place



````python
## read file
df = pd.read_csv("https://raw.githubusercontent.com/sandeepmj/datasets/refs/heads/main/demo-data-solutions/cablebill-v-income.csv")
df.head()
````

````python
## get info
df.info()
````

### Index cable billl

````python
## cable_baseline
cable_baseline = df["avg_cable_bill"].iat[0]
cable_baseline
````

````python
## cable bill index
df["cable_index"] = (
    df["avg_cable_bill"]
    .apply(lambda x: x/cable_baseline * 100)
    .round(1)
)

df.head()
````

### Index median income

````python
## income_baseline
income_baseline = (
    df["median_household_income"].iat[0]
)

income_baseline
````

````python
## median income index
df["median_income_index"] = (
    df["median_household_income"]
    .apply(lambda x: x/income_baseline * 100)
    .round(1)
)

df.head()
````

````python
## analysis
df
````

According to the indexed values, cable prices have rocketed past median household income. The average cable bill in 2023 is 13 times the amount in 1995, while the median household income is only about double the amount across that same span.
